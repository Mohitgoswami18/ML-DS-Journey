## Mathematical operations

```python
x = torch.rand(2,2)
x
```

```python
# addition
x + 2
# substraction
x - 2
# multiplication
x * 3
# division
x / 3
# int division
(x * 100)//3
# mod
((x * 100)//3)%2
# power
x**2
```

```python
a = torch.rand(2,3)
b = torch.rand(2,3)

print(a)
print(b)
```

```python
# add
a + b
# sub
a - b
# multiply
a * b
# division
a / b
# power
a ** b
# mod
a % b
```

```python
c = torch.tensor([1, -2, 3, -4])
```

```python
# abs
torch.abs(c)

# negative
torch.neg(c)

#round
torch.round(c)

#ceil
torch.ceil(c)

# floor
torch.floor(c)

```

```python
e = torch.randint(size=(2,3), low=0, high=10, dtype=torch.float32)
e
```

```python
# sum
torch.sum(e)
# sum along columns
torch.sum(e, dim=0)
# sum along rows
torch.sum(e, dim=1)

# mean
torch.mean(e)
# mean along col
torch.mean(e, dim=0)

# max and min
torch.max(e)
torch.min(e)

# product
torch.prod(e)

# standard deviation
torch.std(e)

# variance
torch.var(e)

# argmax
torch.argmax(e)

# argmin
torch.argmin(e)
```

```python
f = torch.randint(size=(2,3), low=0, high=10)
g = torch.randint(size=(3,2), low=0, high=10)

print(f)
print(g)
```

```python
# matrix multiplcation
torch.matmul(f, g)
```

```python
vector1 = torch.tensor([1, 2])
vector2 = torch.tensor([3, 4])

# dot product
torch.dot(vector1, vector2)
```

```python
# transpose
torch.transpose(f, 0, 1)
```

```python
h = torch.randint(size=(3,3), low=0, high=10, dtype=torch.float32)
h
```

```python
# determinant
torch.det(h)
```

```python
# inverse
torch.inverse(h)
```

```python
i = torch.randint(size=(2,3), low=0, high=10)
j = torch.randint(size=(2,3), low=0, high=10)

print(i)
print(j)
```

```python
# greater than
i > j
# less than
i < j
# equal to
i == j
# not equal to
i != j
# greater than equal to
i >= j
# less than equal to
i <= j
```
