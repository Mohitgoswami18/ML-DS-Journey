## implementing hypothetical testing 

```python
import numpy as np
from scipy.stats import norm
```

```python
# This is only a sample data
# Z test
data = np.array([171, 172, 171, 170, 190])
data_mean = np.mean(data)
pop_mean = 170
pop_sd = 10
alpha = 0.05
z_value = (data_mean - pop_mean) / (pop_sd / np.sqrt(data.shape[0]))
p_value = 2 * ( 1 - norm.cdf(z_value))

if(p_value < alpha) :
    print("Null Hypothesis rejected")
else :
    print("Null Hypothesis accepted")
```

```python
from scipy import stats

data = np.array([171, 172, 171, 170, 190])
pop_mean = 170
alpha = 0.05

t_stat, p_val = stats.ttest_1samp(data, pop_mean)

if p_val < alpha:
    print("Reject Null Hypothesis (mean differs from population mean)")
else:
    print("Fail to Reject Null Hypothesis (no significant difference)")
```

```python
# Chi-Square test
import numpy as np
from scipy.stats import chi2_contingency

data = np.array([[30, 10],
                 [20, 40]])

chi2, p_val, dof, expected = chi2_contingency(data)

alpha = 0.05
if p_val < alpha:
    print("Reject Null Hypothesis (variables are dependent)")
else:
    print("Fail to Reject Null Hypothesis (variables are independent)")
```

```python
from scipy import stats

# Annova test
group1 = [89, 92, 88, 91, 87]
group2 = [78, 75, 77, 79, 76]
group3 = [84, 86, 83, 85, 82]

f_stat, p_val = stats.f_oneway(group1, group2, group3)

alpha = 0.05
if p_val < alpha:
    print("Reject Null Hypothesis (at least one group mean differs)")
else:
    print("Fail to Reject Null Hypothesis (all group means are equal)")
```
