```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
import nltk
from nltk.corpus import stopwords
```

```python
df = pd.read_csv("UpdatedResumeDataSet.csv")
df.head()
```

```python
df.info()
```

```python
def remove_number(text):
    cleaned_text = ""
    for ch in text: 
        if not ch.isdigit(): 
            cleaned_text += ch

    return cleaned_text

df['Resume'] = df['Resume'].apply(remove_number)

df.head()
```

```python
def remove_emoji (text):
    cleaned_text = ""
    for i in text: 
        if i.isascii(): 
            cleaned_text += i
    
    return cleaned_text

df['Resume'] = df['Resume'].apply(remove_emoji)
df.head()
```

```python
nltk.download('stopwords')
```

```python
stop_words = set(stopwords.words('english'))
print(stop_words)
```

```python
def remove_stopwords (text):
    cleaned_text = []
    tokenized = text.split()

    for ch in tokenized: 
        if ch not in stop_words: 
            cleaned_text.append(ch)

    return ' '.join(cleaned_text)

df['Resume'] = df['Resume'].apply(remove_stopwords)
df.head()
```

```python
category = df['Category'].unique()
```

```python
i = 0
category_to_number = {}
for cat in category: 
    category_to_number[cat] = i
    i+= 1

category_to_number
```

```python
df['Category'] = df['Category'].map(category_to_number)
df['Category']
```

```python
vectorizer = CountVectorizer()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df['Resume'], df['Category'], train_size=0.8, random_state=42)
X_train_vectorized = vectorizer.fit_transform(X_train)
X_test_vectorized = vectorizer.transform(X_test)
```

```python
models = {'LR': LogisticRegression(), 'NB': MultinomialNB(), 'SVM': SVC()}
accuracy = {}

for name, model in models.items(): 
    model.fit(X_train_vectorized, y_train)
    y_pred = model.predict(X_test_vectorized)
    acc = accuracy_score(y_pred, y_test)

    accuracy[name] = acc

print(accuracy)
```

```python
from sklearn.model_selection import cross_val_score
for name, model in models.items(): 
    acc = cross_val_score(model, vectorizer.fit_transform(df['Resume']), df['Category'], cv=5)
    print(np.mean(acc))
```
