### Matplotlib -> pythons data visualization library

Matplotlib is a library for creating graphs, charts, visuals very easily

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Creating graphs

```python
x = np.linspace(0,11, 10)
y = x**2

plt.plot(x, y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("First Graph")
```

```python
# histogram 
x = np.arange(20)
y = np.linspace(0,21, 10)
plt.hist(x, y)
```

### Sub Plot

```python
x = np.linspace(1,11,10)
y = x**3
plt.subplot(2,2,1)
plt.plot(x, y)
plt.subplot(2,2,2)
plt.scatter(x, y)
plt.subplot(2,2,3)
plt.bar(x, y)
plt.subplot(2,2,4)
plt.stem(x, y)
```

### Saving a plot

```python
plt.plot(x, y, label="squares")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title("Basic chart")
plt.legend()
plt.grid(visible=True,)
plt.savefig("sqaure_test_graph.png")
plt.show()
```
