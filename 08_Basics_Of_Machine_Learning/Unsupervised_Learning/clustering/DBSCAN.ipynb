```python
from sklearn.datasets import make_moons
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import seaborn as sns
import pandas as pd
```

```python
X, y_true = make_moons(n_samples=500, noise=0.05, random_state=42)
df = pd.DataFrame(X, columns=["Feature1", "Feature2"])
df.head()
```

```python
scale = StandardScaler()
x_scaled= scale.fit_transform(df)
```

```python

model = KMeans(n_clusters=2, random_state=42)
df["labels"] = model.fit_predict(x_scaled)
df.head()
```

```python
sns.scatterplot(data=df, x="Feature1", y="Feature2", hue="labels")
```

```python
df.head()
```

```python
from sklearn.cluster import DBSCAN
model = DBSCAN(eps=0.3, min_samples=5)
model.fit(x_scaled)

df['dbLabels'] = model.labels_
```

```python
df.head()
```

```python
sns.scatterplot(data=df, x="Feature1", y="Feature2", hue="dbLabels")
```
