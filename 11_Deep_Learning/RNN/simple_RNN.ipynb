```python
import numpy as np
from keras import Sequential
from keras.layers import SimpleRNN, Dense
```

```python
model = Sequential()
model.add(SimpleRNN(3, input_shape = (4, 5)))
model.add(Dense(1, activation='sigmoid'))
```
