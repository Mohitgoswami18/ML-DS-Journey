### Pipeline of NLP

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('train.txt', sep=';', names=['text', 'emotion'])
```

```python
df.head()
```

## 1. Lowercasing the text

```python
def lowercase (text):
    return text.lower()

df['text'] = df['text'].apply(lowercase)
df.head(1)
```

## Converting the emotions into numberical values

```python
unique_emotion = df['emotion'].unique()
emotion_to_number = {}
i = 0
for emotion in unique_emotion: 
    emotion_to_number[emotion] = i
    i += 1

print(emotion_to_number)
```

```python
df['emotion'] = df['emotion'].map(emotion_to_number)
df.head()
```

## Removing the punctuations

```python
import string
def remove_punctuations(text):
    return text.translate(str.maketrans('', '', string.punctuation))

df['text'] = df['text'].apply(remove_punctuations)
df.head(2)
```

## Remove Numbers

```python
def remove_number(text):
    cleaned_text = ""
    for ch in text: 
        if not ch.isdigit(): 
            cleaned_text += ch

    return cleaned_text

df['text'] = df['text'].apply(remove_number)

df.head()
```

```python
def remove_emoji (text):
    cleaned_text = ""
    for i in text: 
        if i.isascii(): 
            cleaned_text += i
    
    return cleaned_text

df['text'] = df['text'].apply(remove_emoji)
df.head()
```

## Remove stop words

```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

```python
stop_words = set(stopwords.words('english'))
print(stop_words)
```

```python
def remove_stopwords (text):
    cleaned_text = []
    tokenized = text.split()

    for ch in tokenized: 
        if ch not in stop_words: 
            cleaned_text.append(ch)

    return ' '.join(cleaned_text)

df['text'] = df['text'].apply(remove_stopwords)
df.head()
```
