## Seaborn

Seaborn is a data visualization library built on top of Matplotlib.

It makes it easier to create complex and attractive plots using simple commands.

It works directly with Pandas DataFrames.

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

sns.set_style('dark')
```

Seaborn have inbuilt datasets to work upon

```python
tips = sns.load_dataset('tips')
tips.head(3)
```

Overall there are 2 types of plots 
1. Quantitative Plots
2. Qualitative Plots

### Quantitative

```python
#  1. histplot / distplot -> dist will be depreciated in later versions
sns.histplot(tips['total_bill'], bins=30, kde=True)
```

```python
sns.scatterplot(x=tips['total_bill'], y=tips['tip'], hue=tips['sex'])
```

```python
sns.lineplot(x=tips['total_bill'], y=tips['tip'], hue=tips['sex'])
```

```python
sns.violinplot(x=tips['size'], y=tips['total_bill'])
```

```python
sns.pairplot(tips, kind='reg')
sns.set_style('dark')
```

```python
sns.jointplot(x=tips['total_bill'], y=tips['tip'], kind='reg')
```

### Qualitative

```python
sns.barplot(x=tips['sex'], y=tips['total_bill'])
```

```python
sns.countplot(x='smoker', data=tips)
```

### MATRIX PLOT

```python
pivot_data = tips.pivot_table(values='total_bill', columns='sex', index='tip')
sns.heatmap(pivot_data)
```

```python
pivot = tips.pivot_table(values="tip", index="day", columns="sex", aggfunc="mean")
sns.heatmap(pivot)
```

### Using Seaborn With Matplotlib

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex")
plt.title("Total Bill vs Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.grid(True)
plt.savefig('sample.png')
plt.show()
```
